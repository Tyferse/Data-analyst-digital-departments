```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
vehicles_df = pd.read_csv('vehicles_dataset2.csv', index_col=0)
vehicles_df.head()
```

```python
# Исследование на выбросы
vehicles_df.describe()
```

```python
vehicles_df.odometer = vehicles_df.odometer.fillna(vehicles_df.odometer.mean())
vehicles_df.lat = vehicles_df.lat.fillna(vehicles_df.lat.mean())
vehicles_df.long = vehicles_df.long.fillna(vehicles_df.long.mean())
vehicles_df.describe()
```

```python
plt.hist(vehicles_df.odometer, bins=20)
plt.xlabel('Пробег')
plt.title('Число машин с заданной величиной пробега');
```

```python
plt.boxplot(vehicles_df.odometer)
plt.ylabel('Пробег')
plt.title('Распределние пробега автомобилей')
```

```python
def replace_outliers(column):
    q1, q3 = column.quantile(0.25), column.quantile(0.75)
    iq = q3 - q1
    if pd.api.types.is_integer_dtype(column):
        lb, rb = max(round(q1 - 1.5 * iq), column.min()), min(round(q3 + 1.5 * iq), column.max())
    else:
        lb, rb = max(q1 - 1.5 * iq, column.min()), min(q3 + 1.5 * iq, column.max())

    column[(column < lb)] = lb
    column[(column > rb)] = rb
    return column

vehicles_df.odometer = replace_outliers(vehicles_df.odometer)
plt.boxplot(vehicles_df.odometer)
plt.ylabel('Пробег')
plt.title('Распределение пробега автомобилей после устранения выбросов')
```

```python
plt.boxplot(vehicles_df.price)
plt.ylabel('Цена авто')
plt.title('Распределние цены автомобилей');
```

```python
vehicles_df.price = replace_outliers(vehicles_df.price)
plt.boxplot(vehicles_df.price)
plt.ylabel('Цена авто')
plt.title('Распределение цен автомобилей после устранения выбросов');
```

```python
# Графики категориальных переменных
vehicles_df[['region', 'manufacturer', 'model', 'fuel', 'title_status', 'transmission', 'state', 'price_category']].describe()
```

```python
manufacturers = vehicles_df.groupby('manufacturer', as_index=False).id.count().rename(columns={'id': 'amount'})

# Выделяем отдельно производителей автомобилей, чья доля меньше 2%
other_amount = manufacturers[manufacturers.amount / manufacturers.amount.sum() <= 0.015].amount.sum()
manufacturers = manufacturers[manufacturers.amount / manufacturers.amount.sum() > 0.015]
manufacturers = pd.concat([manufacturers, pd.DataFrame({'manufacturer': ['other'], 'amount': [other_amount]})]).sort_values('amount')

plt.figure(figsize=(6, 10))
plt.pie(manufacturers.amount, labels=manufacturers.manufacturer, autopct='%1.0f%%', labeldistance=None)
plt.legend(loc='upper left', bbox_to_anchor=(0.9, 1))
plt.title('Доля машин производителя');
```

```python
# Зависимости
cols = vehicles_df.corr(numeric_only=True).columns
plt.imshow(vehicles_df.corr(numeric_only=True))
plt.xticks(range(len(cols)), labels=cols)
plt.yticks(range(len(cols)), labels=cols)
plt.colorbar()
plt.title('Корреляция числовых признаков');
```

```python
# Так как значимая корреляция есть только между перемеными price и year, price и odometer и year и odometer,
# то будем строить графики для них

# year_price = vehicles_df.groupby('year', as_index=False).price.mean()
plt.scatter(vehicles_df.year, vehicles_df.price)
plt.xlabel('Год выпуска автомобиля')
plt.ylabel('Стоимость автомобилей')
plt.title('График зависимости цены от года выпуска');
```

```python
colors = {'low': 'blue', 'medium': 'green', 'high': 'red'}
for price_cat in colors:
    data = vehicles_df[vehicles_df.price_category == price_cat]
    plt.hist(data.odometer, bins=20, label=price_cat, color=colors[price_cat], alpha=0.8)

plt.xlabel('Пробег')
plt.ylabel('Количество автомобилей')
plt.title('Распределение пробега по категориям цены')
plt.legend();
```
