## Data Preparation

### Загрузка данных

```python
df = pd.read_csv('df_out.csv')
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], utc=True)
```

### Обработка пропусков

#### Числовые признаки

```python
df.describe()
```

```python
# Заменяем на медиану, потому что на среднее значение оказывают влияние те, которые ниже 25% квантиля
df['year'] = df['year'].fillna(df['year'].median()).astype(int)
```

```python
# Проверим, чем можно заполнить пропуски координат. 
# Предполодим, есть связь между штатом или регионом и координатами
region_coords = pd.concat([df.groupby('region')[['lat']].mean().rename(columns={'lat': 'lat_mean'}), 
                           df.groupby('region')[['lat']].std().rename(columns={'lat': 'lat_std'}),
                           df.groupby('region')[['long']].mean().rename(columns={'long': 'long_mean'}), 
                           df.groupby('region')[['long']].std().rename(columns={'long': 'long_std'})], axis=1)
region_coords[['lat_std', 'long_std']].mean()
```

```python
# Теперь сгруппируем по штатам
state_coords = pd.concat([df.groupby('state')[['lat']].mean().rename(columns={'lat': 'lat_mean'}), 
                          df.groupby('state')[['lat']].std().rename(columns={'lat': 'lat_std'}),
                          df.groupby('state')[['long']].mean().rename(columns={'long': 'long_mean'}), 
                          df.groupby('state')[['long']].std().rename(columns={'long': 'long_std'})], axis=1)
state_coords[['lat_std', 'long_std']].mean()
```

```python
# Так как и штаты, и регионы известны (в них нет пропусков), 
# то подставляем средние координаты по регионам, т. к. у них меньшая дисперсия
df.loc[df['lat'].isna() | df['long'].isna(), 'lat'] = df[df['lat'].isna() | df['long'].isna()].apply(lambda x: region_coords.loc[x.region].lat_mean, axis=1)
df.loc[df['lat'].isna() | df['long'].isna(), 'long'] = df[df['lat'].isna() | df['long'].isna()].apply(lambda x: region_coords.loc[x.region].long_mean, axis=1)

df.describe()
```

#### Категориальные признаки

```python
df.describe(include='object')
```

```python
# Заполняем колонки fuel, title_status и transmission модой 
# так как у них мало уникальных значений и отностильно небольшое количество пропусков
df['fuel'] = df['fuel'].fillna(df['fuel'].mode()[0])
df['title_status'] = df['title_status'].fillna(df['title_status'].mode()[0])
df['transmission'] = df['transmission'].fillna(df['transmission'].mode()[0])
```

```python
# Находим в описании модель по производителю, и если найдена, заполняем ей
def model_in_desc(desc, mnf):
    global df

    if mnf is None:
        return

    models = df[(df['manufacturer'] == mnf) & (~df['model'].isna())].model.unique()
    for model in models:
        if model in desc:
            return model

df.loc[(~df.description.isna()) & df.model.isna(), 'model'] = (df[(~df.description.isna()) & df.model.isna()]
                                                               .apply(lambda x: model_in_desc(x.description, x.manufacturer), axis=1))
df['model'].isna().sum()
```

```python
# Остальные модели заменяем как самые популярные по производителю
df.loc[df.model.isna(), 'model'] = df[df.model.isna()].apply(lambda x: df[df['manufacturer'] == x.manufacturer].model.mode()[0], axis=1)
df['model'].isna().sum()
```

```python
df.describe(include='object')
```

### Устранение аномалий

Проверяем year, price и odometer на выбросы и заменяем их на граничные значения ящиков с усами

```python
df.year.plot.box(title='Распределение годов выпуска автомобилей');
```

```python
q1, q3 = df.year.quantile(0.25), df.year.quantile(0.75)
lb = ceil(q1 - 1.5 * (q3 - q1))
df.loc[df['year'] < lb, 'year'] = lb
df['year'] = df['year'].astype(int)

df.year.plot.box(title='Распределение годов выпуска автомобилей\nпосле устранения выбросов');
```

```python
df.price.plot.box(title='Распределение цен автомобилей');
```

```python
q1, q3 = df.price.quantile(0.25), df.price.quantile(0.75)
lb, rb = max(q1 - 1.5 * (q3 - q1), df.price.min()), min(q3 + 1.5 * (q3 - q1), df.price.max()) 
df.loc[df['price'] < lb, 'price'] = lb
df.loc[df['price'] > rb, 'price'] = rb

df.price.plot.box(title='Распределение цен автомобилей\nпосле устранения выбросов');
```

```python
# Смотрим на график и видим, что выбросов нет
df.odometer.plot.box(title='Распределение пробега автомобилей');
```

## Feature engineering

### Создание новых признаков

```python
# наличие названия модели в описании
df['model_in_desc'] = df.apply(lambda x: int(x.model in x.description) if isinstance(x.description, str) else 0, axis=1)

# Количество слов в тексте описания
df['desc_words'] = df.description.apply(lambda x: len(x.split(' ')) if isinstance(x, str) else 0)
```

```python
# Выделяем признаки из дат
df['pub_day'] = df.date.dt.day.fillna(df.date.dt.day.mode()[0]).astype(int)
df['pub_month'] = df.date.dt.month.fillna(df.date.dt.month.mode()[0]).astype(int)
df['pub_year'] = df.date.dt.year.fillna(df.date.dt.year.mode()[0]).astype(int)
df['pub_day_of_week'] = df.date.dt.weekday
df['pub_day_of_week'] = df['pub_day_of_week'].fillna(df['pub_day_of_week'].mode()[0]).astype(int)

print(df['pub_month'].value_counts(), df['pub_year'].value_counts(), sep='\n\n')
```

```python
# Так как год всего один, а месяцев, то pub_year удаляем, а pub_month заменяем на 1 и 0
df = df.drop('pub_year', axis=1)
df['pub_month'] = df['pub_month'].apply(lambda x: int(x == 4))

df['pub_month'].value_counts()
```

```python
# Разница в годах между годом публикации и годом выпуска
df['pub_year_diff'] = (2021 - df.year).astype(int)
```

### Кодирование категориальных переменных

```python
ohe = OneHotEncoder(sparse_output=False)
category_labels = ['region', 'manufacturer', 'model', 'fuel', 'title_status', 'transmission', 'state']
encoded_categories = ohe.fit_transform(df[category_labels]) 
encoded_categories
```

```python
df[ohe.get_feature_names_out()] = encoded_categories.astype(int)
df = df.drop(category_labels, axis=1)
df.head()
```

### Стандартизация количественных признаков

```python
scaler = StandardScaler()
numeric_columns = ['price', 'year', 'odometer', 'lat', 'long', 'desc_words', 'pub_day', 'pub_day_of_week', 'pub_year_diff']
scaled_numeric = scaler.fit_transform(df[numeric_columns])
scaled_numeric
```

```python
df[numeric_columns] = scaled_numeric
df[numeric_columns].head()
```

```python
df.iloc[:, :10].isna().sum()
```

### Сохранение датасета

```python
df = df.drop(['id', 'url', 'region_url', 'price', 'image_url', 'description', 'posting_date', 'date'], axis=1)
df.head()
```

## Modeling

### Обучение моделей

```python
x = df.drop('price_category', axis=1)
y = df['price_category']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
logreg = LogisticRegression()
randforc = RandomForestClassifier()
mlperc = MLPClassifier()
```

```python
param_grid = {
    'penalty': ['l2', 'l1'], # 'elasticnet',
    # 'C': [10**i for i in range(-3, 3)],
    'solver': ['newton-cg', 'lbfgs', 'saga'],
    # 'l1_ratio': [i / 4 for i in range(1, 4)]
    'n_jobs': [-1],
}

rs_logreg = RandomizedSearchCV(
    estimator=logreg,
    param_distributions=param_grid,
    scoring='accuracy',
    n_iter=20,
    verbose=1,
    n_jobs=-1
)

rs_logreg.fit(x_train, y_train)

best_params = rs_logreg.best_params_
best_params
```

```python
"""
tol=1e-12, max_iter=1000  0.7324
max_iter=1000   0.7316
solver='newton-cg'  0.7328
penalty='elasticnet', l1_ratio=0.1, solver='saga'   0.7232
C=10, solver='newton-cg'    0.7356
C=5, solver='newton-cg'     0.7364
C=2, solver='newton-cg'     0.738
C=2.5, solver='newton-cg'   0.7384
C=2.4, solver='newton-cg'   0.7384
"""

lgr = LogisticRegression(C=2, solver='newton-cg', n_jobs=-1)
lgr.fit(x_train, y_train)

lgr_pred = lgr.predict(x_test)
accuracy_score(y_test, lgr_pred)
```

```python
confusion_matrix(y_test, lgr_pred)
```

```python
cvs = cross_validate(lgr, x_train, y_train, cv=5, n_jobs=-1)
cvs
```

```python
param_grid = {
    'max_depth': list(range(8, 49, 8)),
    'min_samples_split': list(range(2, 9, 2)),
    'n_estimators': list(range(100, 501, 100))
}

rs_rfс = RandomizedSearchCV(
    estimator=randforc,
    param_distributions=param_grid,
    scoring='accuracy',
    n_iter=20,
    verbose=1,
    n_jobs=-1
)

rs_rfс.fit(x_train, y_train)

best_params = rs_rfс.best_params_
best_params
```

```python
# {'n_estimators': 400, 'max_depth': 48}    0.7296
# {'n_estimators': 400, 'max_depth': 60}    0.736
# {'n_estimators': 400, 'max_depth': 120}   0.744
# {'n_estimators': 400, 'max_depth': 115}   0.748
# {'n_estimators': 425, 'max_depth': 115}   0.7428
rfc = RandomForestClassifier(n_estimators=400, max_depth=56, n_jobs=-1)
rfc.fit(x_train, y_train)

rfc_pred = rfc.predict(x_test)
accuracy_score(y_test, rfc_pred)
```

```python
confusion_matrix(y_test, rfc_pred)
```

```python
cvs_rfc = cross_validate(rfc, x_train, y_train, cv=5, n_jobs=-1)
cvs_rfc
```

```python
# hidden_layer_sizes=[120, 24], learning_rate='adaptive', activation='relu'     0.752
# hidden_layer_sizes=[240, 24], learning_rate='adaptive', activation='relu'     0.7624
# hidden_layer_sizes=[240], learning_rate='adaptive', activation='relu'         0.7572
# hidden_layer_sizes=[240, 24], learning_rate='invscaling', activation='relu'   0.7692
# hidden_layer_sizes=[320, 27], learning_rate='invscaling', activation='relu'   0.7596
# hidden_layer_sizes=[200, 27], learning_rate='invscaling', activation='relu'   0.766
# hidden_layer_sizes=[240, 32], learning_rate='invscaling', activation='relu'   0.77
# hidden_layer_sizes=[120, 16], learning_rate='invscaling', activation='relu'   0.7672
mlp = MLPClassifier(hidden_layer_sizes=[240, 32], learning_rate='adaptive', activation='relu', early_stopping=True)
mlp.fit(x_train, y_train)

mlp_pred = mlp.predict(x_test)
accuracy_score(y_test, mlp_pred)
```

```python
confusion_matrix(y_test, mlp_pred)
```

```python
cvs_mlp = cross_validate(mlp, x_train, y_train, cv=5, n_jobs=-1)
cvs_mlp
```
